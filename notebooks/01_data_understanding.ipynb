## Target Variable Analysis

```python
df["Churn Label"].value_counts()
df["Churn Label"].value_counts(normalize=True) * 100
```

### Observation

Approximately 26.5% of customers have churned, while 73.5% have remained. The dataset is moderately imbalanced, which should be considered during model training and evaluation.

```python
import matplotlib.pyplot as plt

df["Churn Label"].value_counts().plot(kind="bar")
plt.title("Churn Distribution")
plt.xlabel("Churn Label")
plt.ylabel("Number of Customers")
plt.show()
```

```python
pd.crosstab(df["Contract"], df["Churn Label"], normalize="index") * 100
```

### Observation

The analysis shows a strong relationship between contract type and customer churn.

Customers with **month-to-month contracts** have the highest churn rate (**42.71%**), while customers with **two-year contracts** have the lowest churn rate (**2.83%**).

This suggests that longer contract durations significantly improve customer retention. Therefore, **Contract** is expected to be one of the most important predictive features in the churn prediction model.

From a business perspective, telecom companies may reduce customer churn by encouraging customers to switch from month-to-month contracts to longer-term agreements through promotions or loyalty programs.

## Contract Type vs Churn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Analysis Objective

The objective of this analysis is to investigate how different contract types affect customer churn. Understanding this relationship helps identify high-risk customer groups and provides valuable business insights for improving customer retention.

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="Contract",
    hue="Churn Label"
)

plt.title("Customer Churn by Contract Type", fontsize=14)
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")

plt.legend(title="Churn")
plt.tight_layout()

plt.show()
```

### Observation

The visualization clearly shows that customers with **month-to-month contracts** have a much higher churn rate than customers with one-year or two-year contracts. In contrast, customers with long-term contracts are much more likely to remain with the company.

### Business Insight

Customers with month-to-month contracts represent the highest-risk group. Telecom companies can improve customer retention by encouraging these customers to switch to longer-term contracts through loyalty programs, discounts, or promotional offers.

### Machine Learning Perspective

The **Contract** feature is expected to be one of the most influential variables in the churn prediction model because it shows a strong relationship with customer churn.

# Internet Service vs Churn

### Analysis Objective

```python
pd.crosstab(
    df["Internet Service"],
    df["Churn Label"],
    normalize="index"
) * 100
```

### Observation

Customers using **Fiber optic** internet have the highest churn rate (**41.89%**), while customers without internet service have the lowest churn rate (**7.40%**). DSL customers show a moderate churn rate (**18.96%**).

### Business Insight

Fiber optic customers appear to represent a high-risk customer segment. The company should further investigate customer satisfaction, service quality, pricing, and competitor offerings to understand why churn is significantly higher among these customers.

### Machine Learning Perspective

The **Internet Service** feature demonstrates a strong relationship with customer churn and is expected to be an important predictor in the machine learning model.

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="Internet Service",
    hue="Churn Label"
)

plt.title("Customer Churn by Internet Service", fontsize=14)
plt.xlabel("Internet Service")
plt.ylabel("Number of Customers")

plt.legend(title="Churn")
plt.tight_layout()

plt.show()
```

### Observation

Customers using **Fiber optic** internet exhibit the highest churn rate (**41.89%**), while customers without internet service have the lowest churn rate (**7.40%**). DSL customers show a moderate churn rate (**18.96%**).

### Business Insight

The company should investigate why Fiber optic customers are leaving at a significantly higher rate. Possible factors such as pricing, service quality, customer support, or competitor offers should be explored in future analyses.

### Machine Learning Perspective

The **Internet Service** feature shows a strong relationship with customer churn and is expected to be one of the key predictive variables in the machine learning model.
